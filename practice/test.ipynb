## Visualizations!

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8')
```

```python
close = stocks.loc[:, 'Close']
close.plot(figsize=(12, 8), fontsize=12)
plt.show()
```

```python
normClose = close / close.iloc[0] * 100
normClose
```

```python
normClose.plot(figsize=(12, 8), fontsize=12)
plt.show()
```

```python
aapl = close.AAPL.copy().to_frame()
aapl
```

```python
aapl['% change'] = aapl.AAPL.pct_change() * 100
aapl
```

```python
aapl['% of prev close'] = aapl.AAPL.shift() / aapl.AAPL
aapl
```

```python
aapl['diff'] = aapl.AAPL.diff()
aapl
```

```python
aapl.AAPL.resample('BME').last().pct_change()
```

```python
aapl = aapl[['AAPL', 'diff', '% change', '% of prev close']]
aapl
```
