```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8')
```

```python
tickers = ['SPY', 'AAPL', 'KO', 'MSFT', 'DIS', 'IBM']
stocks = yf.download(tickers, start='2010-01-01')
stocks
```

```python
close = stocks.Close.copy()
close
```

```python
normClose = close / close.iloc[0] * 100
normClose
```

```python
normClose.plot(figsize=(12, 7))
```

```python
ret = close.pct_change().dropna() * 100
ret
```

```python
ret.describe().T
```

```python
dailySummary = ret.describe().T[['mean', 'std']]
dailySummary
```

```python
annualSummary = dailySummary * [252, np.sqrt(252)]
annualSummary
```

```python
annualSummary.plot.scatter(x='std', y='mean', figsize=(10,6), s=50)

for label in annualSummary.index:
    plt.annotate(label, tuple(annualSummary.loc[label])[::-1], textcoords='offset points', xytext=(0, 6))
plt.xlabel('Annual risk (std)')
plt.ylabel('Annual return (mean)')
plt.title('Risk/return', fontsize=20)
plt.show()
```
